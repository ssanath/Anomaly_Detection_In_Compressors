```python
z_scores = np.abs((df3['CFM'] - df3['CFM'].mean()) / df3['CFM'].std())

threshold = 2.5

outliers = df3[z_scores > threshold]
non_outliers = df3[z_scores <= threshold]
```

```python
plt.scatter(non_outliers['ACTIVEPOWER'], non_outliers['CFM'], c='b', label='Normal Data', s=10)
plt.scatter(outliers['ACTIVEPOWER'], outliers['CFM'], c='r', label='Outliers', s=20, marker='x')
plt.show() # plot outliers
```

```python
non_outliers.head()
```

```python
outliers.info()
```

```python
non_outliers.info()
```

```python
X_train = np.r_[non_outliers, outliers]
```

```python
X_test = np.r_[non_outliers]
```

```python
model = svm.OneClassSVM(kernel='rbf', gamma=0.3) #use rbf kernel, changing gamma value to 0.3 improved the accuracy slightly
model.fit(X_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(np.ones(len(X_test)), y_pred) # predict accuracy where the first parameter is an array of ones, all the data points belong to the same class i.e; none are outliers. This is compared with the predicted data points.
print(f"Accuracy: {accuracy}")
```

```python
precision = precision_score(np.ones(len(X_test)), y_pred)
print(f"Precision: {precision}")
```

```python
from sklearn.metrics import recall_score
recall = recall_score(np.ones(len(X_test)), y_pred)
print(f"Recall: {recall}")
```

```python
#yellow points were correctly predicted and purple points were predicted as anomalies
plt.scatter(X_test[:, 1], X_test[:, 0], c=y_pred, cmap='viridis')
plt.title("One-Class SVM Anomaly Detection")
plt.xlabel("AP")
plt.ylabel("CFM")
plt.colorbar()
plt.show()
```
